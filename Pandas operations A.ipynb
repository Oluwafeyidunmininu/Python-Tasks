#Creating A Table and Visuals on Panda 

```python
#Converting to dataframe

import pandas as pd
```

```python
data = {'Name': ['John', 'Jane', 'Jim', 'Jake'], 'Age':[28,24,33,29], 'Occupation': ['Engineer', 'Doctor', 'Artist', 'Scientist']}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
#Merging and Joining DataFrames
#Concatenation
df1 = pd.DataFrame({'Name': ['Ade', 'Biola'], 'Age': [23, 24]})
df2 = pd.DataFrame({'Name': ['Caro', 'Dammy'], 'Age': [25, 26]})
pd.concat([df1, df2],ignore_index=True) 
```

```python
#Merging 
df1= pd.DataFrame({'PersnonalID': [1,2,3], 'First_Name':['Ruth','Deji','Joy']})
df2= pd.DataFrame({'PersnonalID': [1,2,3], 'Salary':[10000, 20000, 30000]})
df_merged = pd.merge(df1,df2, on='PersnonalID', how = 'inner') 
df_merged
```
