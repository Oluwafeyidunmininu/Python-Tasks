```python
from sklearn.datasets import load_iris
```

```python
#loading the iris dataset 
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
#scattterPlot (Petal_length vs Petal_width)
sns.scatterplot(x = 'petal_length', y = 'petal_width', data = iris )
```

```python
iris.info()
```

```python
#Box plot (sepal_length by species)
sns.boxplot( data =iris, x='sepal_length', y = 'species')
```

```python
#Pair plot for relationships
sns.pairplot(iris)
```

```python
#2 Heatmap
# Load Iris dataset
iris = load_iris()

# Creating a DataFrame from the dataset
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Creating a heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(iris_df.corr(), annot=True, cmap="coolwarm", square=True)
```

```python
#Violin plot
sns.violinplot(iris)
```
